```python
import os
import sys
import pandas as pd
import numpy as np
os.chdir('C:\\Users\\sweet\\data')

# Filter out the leased vehicles
for fname in os.listdir(os.getcwd()):
    if 'Leas' in fname:
        os.remove(os.path.join(os.getcwd(), fname))
```

```python
# create a dataframe to store the merged data
df = pd.DataFrame()
```

```python
for fname in os.listdir(os.getcwd()):
    data = pd.read_csv(fname, dtype={"assetNumber": "string"}, low_memory=False)
    # drop all columns but assetNumber, originatorName, originationDate, and PTI
    data = data[['assetNumber', 'originatorName', 'originationDate', 'paymentToIncomePercentage']]
    df = df.append(data)
```

```python
# keep only one copy of duplicate rows, except for all nulls and save to file
df = df[(~df.duplicated()) | (df['assetNumber'].isnull())]
df.to_csv("../merged.csv")
```

```python
# drop only invalid null
df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
# convert pti strings to numeric
df['paymentToIncomePercentage'] = df['paymentToIncomePercentage'].str.extract('(\d*\.\d+|\d+)', expand=False).astype(float)
```

```python
df
```

```python
df['paymentToIncomePercentage'] = np.where(df['paymentToIncomePercentage'] < 1, df['paymentToIncomePercentage'], df['paymentToIncomePercentage']/100)
```

```python
df['originationDate'] = pd.to_datetime(df['originationDate'])
```
